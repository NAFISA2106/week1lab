Problem 1


```
scores = [72 , 85 , 90 , 68 , 88]
print("Scores:", scores)

average = sum(scores) / len(scores)
print("Average score:", average)

highest_score = max(scores)
print("Highest score:", highest_score)
lowest_score = min(scores)
print("Lowest score:", lowest_score)
```

Problem 3

```
scores = [45 , 67 , 82 , 59]

def pass_fail(score):
  if score >= 60:
    return "Pass"
  else:
    return "Fail"

for score in scores:
  print("Score:", score, "->", pass_fail(score))
```

Problem 4

```
from re import T
sizes = [500 , 800 , 1000 , 1200]
prices = [50 , 80 , 100 , 130]

for i in range(len(sizes)):
  print("Sizes:", sizes[i], "sq.m -> Price:", prices[i])

is_incresing = True
for i in range(1, len(prices)):
  if prices[i] <= prices[i - 1]:
    is_incresing = False

if is_incresing:
  print("Prices trend")
else:
  print("Negative trend")
```

Problem 5

```
study_hours = [2 , 4 , 6 , 8]
scores = [55 , 65 , 75 , 85]

increase = (scores[-1] - scores[0]) / (study_hours[-1] - study_hours[0])
predicted_score = scores[-1] +  increase * 2

print("Average increase per 2 hours:", increase * 2)
print("Predicted score for 10 hours:", predicted_score)
```

Problem 6

```
scores = [35 , 58 , 62 , 74 , 89 , 45]

for s in scores:
  if s < 50:
    category = "Low"
  elif s < 70:
    category = "Medium"
  else:
    category = "High"
print(s, "->", category)
```

Problem 7

```
ages = [18 , 20 , 22 , 24 , 26]
heights = [160 , 165 , 170 , 175 , 180]

for a, h in zip(ages, heights):
  print(f"Age: {a} -> Height: {h}")

print("Average age:" , sum(ages) / len(ages))
print("Average height:" , sum(heights) / len(heights))
print("Hight increases as age increases")
```
